```python
import pandas as pd

def unistats(df):
    output_df = pd.DataFrame(columns=['Count', 'Missing', 'Unique', 'Dtype', 'Numeric', 'Mean', 'Mode', '25%', 'Min', '75%', 'Max', 'Std', 'Skew', 'Kurt'])
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            numeric = True
        else:
            numeric = False
        
        if numeric:
            output_df.loc[col] = [df[col].count(), df[col].isnull().sum(), df[col].nunique(), df[col].dtype, numeric, df[col].mean(), df[col].mode().values[0], df[col].quantile(0.25), df[col].min(), df[col].quantile(0.75), df[col].max(), df[col].std(), df[col].skew(), df[col].kurt()]
        else:
            output_df.loc[col] = [df[col].count(), df[col].isnull().sum(), df[col].nunique(), df[col].dtype, numeric, '', df[col].mode().values[0], '', '', '', '', '', '', '']
    return output_df.sort_values(by=['Numeric','Skew', 'Unique'], ascending = False)

# Test the function
import pandas as pd
df = pd.read_csv('housing_full.csv')
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
unistats(df)
```

```python
def bivstats(df, label):
    from scipy import stats
    import numpy as np

    # Create an empty DataFrame to store the correlation and p-values
    output_df = pd.DataFrame(columns=['r', 'F', 'X2', 'p-value'])

    for col in df.columns:
        if not col == label:
            if df[col].isnull().sum() == 0:
                if pd.api.types.is_numeric_dtype(df[col]):  # Only calculate r and p-value for numeric columns
                    r, p = stats.pearsonr(df[label], df[col])
                    output_df.loc[col] = [round(r, 3), '-', '-', round(p, 6)]
        else:
            output_df.loc[col] = [np.nan, '-', '-', 'nulls']

    # Sort by the absolute values of the correlation coefficients
    output_df['abs_r'] = output_df['r'].abs()  # Add a new column for absolute values of r
    output_df = output_df.sort_values(by='abs_r', ascending=False).drop(columns=['abs_r'])

    return output_df

import pandas as pd
df = pd.read_csv('housing_full.csv')
result = bivstats(df, 'SalePrice')
print(result)
```
